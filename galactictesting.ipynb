```python
import torch, os 
from ultralytics import YOLO
from pathlib import Path
```

```python
modelpath = Path(r"C:\Users\ahmed\Desktop\me\me\Coding\HARDCORE\Galatic_Defender\galactic.pt")
model = YOLO(modelpath)
```

```python
import cv2 , time
cap = cv2.VideoCapture(Path(r"C:\Users\ahmed\Desktop\api_thumb_600.mp4"))

fps_times = []
while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    t0 = time.perf_counter()
    results = model.track(frame, conf=0.25, verbose=False,persist=True)
    ms = (time.perf_counter() - t0) * 1000
    fps_times.append(ms)
    annotated = results[0].plot()
    cv2.imshow("YOLO26n — Anti-UAV", annotated)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

cap.release(); cv2.destroyAllWindows()
avg_ms = sum(fps_times)/len(fps_times)
print(f"Avg inference: {avg_ms:.1f}ms | Theoretical FPS: {1000/avg_ms:.0f}")
```
